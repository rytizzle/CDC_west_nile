```python
import pandas as pd

weather = pd.read_csv('./assets/input/weather.csv')

train = pd.read_csv('./assets/input/train.csv')

train['Date'] = pd.to_datetime(train['Date'])

weather['Date'] = pd.to_datetime(weather['Date'])
```

```python
print(weather.shape)
print(train.shape)
```

```python
weather_station1 = weather.drop_duplicates('Date')

weather['Station_2'] = pd.get_dummies(weather['Station'], drop_first= True)


df = train.merge(weather_station1, how = 'inner', on = 'Date', copy = False)
```

```python
df.columns
```

```python
df
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

# Create X
### Make dummy variables of categorical

```python
#mosquito species dummies
mos_dum = pd.get_dummies(df['Species'], drop_first= True)

mos_dum
```

```python
X_o = df.loc[:, ['SeaLevel', 'Tavg', 'AvgSpeed', 'DewPoint', 'NumMosquitos']]
```

```python
X = df.loc[:, ['SeaLevel', 'Tavg', 'AvgSpeed', 'DewPoint', 'NumMosquitos']]
```

```python
X = X.join(mos_dum)
```

```python
y = df['WnvPresent']
```

```python
X['SeaLevel'] = pd.to_numeric(X.SeaLevel)
X['Tavg'] = pd.to_numeric(X.Tavg)
X['AvgSpeed'] = pd.to_numeric(X.AvgSpeed)
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```
